```python
# Question 1 (Marks - 1): Using the TIPS dataset, Create a pivot_table showing average tip amount spent each day for each gender
# ANSWER 1

import pandas as pd
tips='tips.csv'
tips=pd.read_csv(tips)
tips
import numpy as np
table = pd.pivot_table(tips,index=['day'],columns=['sex'],aggfunc={'tip':np.mean})
table
```

```python
# Question 2 (Marks â€“ 1 ): Using the TITANIC dataset, use loc and Boolean indexing to return the data frame only for women who did not survive the crash and were travelling in the third class.Print the last 5 data rows
# ANSWER 2 
Titanic = 'TITANIC(1).csv'
Titanic = pd.read_csv(Titanic)
Titanic
Passenger_details = Titanic.loc[(Titanic['Sex'] == 'female') & (Titanic['Survived'] == 0) &   (Titanic['Pclass'] == 3)]
LAST_ROWS = Passenger_details.tail(5)
LAST_ROWS
```

```python
# Question 3 (Marks â€“ 1 ) Sort the TIPS data frame with lowest to highest Tip Percentage (Hint: You need to first create a Tip Percentage column).
# Answer 3

TIPS = pd.DataFrame(tips)
TIPS['tip_percentage'] = (TIPS['tip']/TIPS['total_bill'])*100
TIPS
SORTING_TIPS_DF = TIPS.sort_values(by='tip_percentage',ascending=True)
SORTING_TIPS_DF
```

```python
# Question 4 (Marks â€“ 1 ) Using the TITANIC dataset, on average how much do male smokers spend in the restaurant for their food bill (excluding tips)?
# ANSWER 4

MALE_SMOKER_DETAILS = tips[(tips['sex']=='Male') & (tips['smoker'] == 'Yes')]
Average_Spent = (MALE_SMOKER_DETAILS['total_bill'] - MALE_SMOKER_DETAILS['tip']).mean()
Average_Spent 
```

```python
# Question 5a (Marks â€“ 0.5 ) Using the TITANIC dataset. Calculate the mean , and median age for each Pclass (Use .groupby())Expected.
# ANSWER 5a
MEAN_MEDIAN_AGE_PASSENEGERS = Titanic.groupby('Pclass')['Age'].agg(['median','mean']).reset_index()
MEAN_MEDIAN_AGE_PASSENEGERS 
```

```python
# Question 5b (Marks â€“ 0.5 ) Using the TITANIC dataset. Calculate the mean , and median age for each Pclass (Use .pivot_table())
# ANSWER 5b
Age_Pivot_table = Titanic.pivot_table(index='Pclass',values='Age', 
aggfunc={'Age':['mean','median']})
Age_Pivot_table.columns = ['MEAN-AGE','MEDIAN-AGE']
Age_Pivot_table

```
